```python
import numpy as np 
import matplotlib.pyplot as plt 

from scipy import random 
from scipy.stats import multivariate_normal as mvn 

from main import GMM 

%matplotlib inline
```

```python
def generate_data(k, dim=2, N=200, lim=[-10, 10]): 
    """Generate data from a radom mixture of Gaussians in a given range.
    
    Plot the points in the case of 2D.
    
    Input: 
    - k: the number of clusters.
    - dim: dimension of generated points. 
    - N: the number of points to be generated for each cluster.
    - lim: range of mean values.
    """
    
    X = []    # data matrix 
    
    # generate the mean values 
    mean = random.rand(k, dim) * (lim[1] - lim[0]) + lim[0]
    
    # generate covariance matrices for each cluster 
    for i in range(k): 
        cov = random.rand(dim, dim + 10)
        cov = np.matmul(cov, cov.T)
        
        x = np.random.multivariate_normal(mean[i], cov, N)    # generate random samples
        X += list(x)
        
    X = np.array(X)
    
    if dim == 2:
        plt.figure(figsize=(10, 6))
        
        plt.scatter(X[:, 0], X[:, 1], s=3, alpha=0.4)
        plt.show()
        
    return X
```

```python
def plot_2D(X, title=None):
    """Draw the data points and the fitted mixture model.
    
    - X: data matrix 
    - title: title of plot and name with which it will be saved.
    """
    
    fig, ax = plt.figure(figsize=(10, 8))
    
    ax.scatter(X[:, 0], X[:, 1], s=3, alpha=0.4)
    ax.scatter(gmm.mu[:, 0], gmm.mu[:, 1], c=gmm.colors)    # draw normal means 
    
    gmm.draw(ax, lw=3)
    
    ax.set_xlim((-12, 12))
    ax.set_ylim((-12, 12))
    
    plt.title(title)
    plt.show()
    plt.clf()
```

```python
X = generate_data(k=3, dim=2, N=1000)
```

```python
# create a GMM 
gmm = GMM(k=3, dim=2)

# training GMM using EM 
gmm.fit(X)
```
